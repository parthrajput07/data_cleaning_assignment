```python
df_clean = df.copy()
```

```python
df_clean.isnull().sum()
```

```python
df_clean["discount"] = df_clean["discount"].fillna(df_clean["discount"].mean())
```

```python
df_clean.isnull().sum()
```

```python
df_clean["what_customer_said"] = df_clean["what_customer_said"].fillna("No review")
```

```python
df_clean["what_customers_said"] = df_clean["what_customers_said"].fillna("No Reviews")
```

```python
df_clean.isnull().sum()
```

```python
df_clean["seller_name"] = df_clean["seller_name"].fillna("Unknown")
df_clean["videos"] = df_clean["videos"].fillna("No Video")
df_clean["seller_information"] = df_clean["seller_information"].fillna("Not Available")
df_clean["variations"] = df_clean["variations"].fillna("No Variation")
```

```python
df_clean.isnull().sum()
```

```python
df_clean.duplicated.sum()
```

```python
df_clean.duplicated().sum()
```

```python
df_clean.drop_duplicates(inplace=True)
```

```python
df_clean[["title","rating","initial_price"]].head()
```

```python
df_clean[df_clean["rating"] > 4.5]
```

```python
df_clean["discount_amount"] = (
    df_clean["initial_price"] * df_clean["discount"] / 100
)
```

```python
df_clean[["initial_price","discount","discount_amount"]].head()
```

```python
df_clean.to_csv("cleaned_ecommerce_dataset.csv", index=False)
```
